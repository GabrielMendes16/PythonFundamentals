## O que são Tipos de Dados?

Os tipos de dados são categorias que definem o tipo de valor que uma variável pode armazenar. Compreender os tipos de dados é fundamental para manipular informações de maneira eficaz em Python.


## Tipos de Dados Comuns

### Inteiros

Os inteiros são números inteiros, sem parte decimal.

```python
# Célula de código: Exemplos de inteiros
numero_inteiro = 10
print(f"Número inteiro contido na variável numero_inteiro: {numero_inteiro}")

```

### Floats

Os floats são números que contêm uma parte decimal.

```python
# Célula de código: Exemplos de floats
numero_decimal = 3.14
print(f"Número decimal contido na variável numero_decimal: {numero_decimal}")
```

### Strings

As strings são sequências de caracteres, utilizadas para representar texto.

```python
# Célula de código: Exemplos de strings
texto = "Olá, Mundo!"
print(f"Texto contido na variável texto: {texto}")
```

### Booleanos
Os booleanos representam valores de verdade: True ou False.

```python
# Célula de código: Exemplos de booleanos
verdadeiro = True
falso = False
print(f"Verdadeiro: {verdadeiro}, Falso: {falso}")
```

### Listas
As listas são coleções ordenadas e mutáveis de itens.

``` python
# Célula de código: Exemplos de listas
lista = [1, 2, 3, 4, 5]
print(f"Lista: {lista}")
```
### Tuplas
As tuplas são coleções ordenadas e imutáveis de itens.

``` python
# Célula de código: Exemplos de tuplas
tupla = (1, 2, 3)
print(f"Tupla: {tupla}")
``` 

### Dicionários
Os dicionários são coleções de pares chave-valor.

``` python
# Célula de código: Exemplos de dicionários
dicionario = {"nome": "Alice", "idade": 30}
print(f"informações contidas no Dicionário: {dicionario}")
```
# Conversão de Tipos de Dados

É possível converter um tipo de dado em outro usando funções específicas. Veja alguns exemplos:

# Célula de código: Exemplos de conversão
``` python
numero = "123"  # String
numero_convertido = int(numero)  # Converte para inteiro
print(f"Número convertido: {numero_convertido}")

decimal = "3.14"  # String
decimal_convertido = float(decimal)  # Converte para float
print(f"Decimal convertido: {decimal_convertido}")
```

```python
quantidade = 0

    quantidade = quantidade + 1
    print(nomes)

```
